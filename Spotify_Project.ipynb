# Is there a correlation between danceability and energy in the tracks?
Yes. Weak positive correlation(0.1)

```python
df_corr = df.copy()
```

```python

corr = df_corr.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True



# Set up the matplotlib figure
fig, ax = plt.subplots(figsize=(12, 9))

# Draw the heatmap with 'sns.heatmap()'
ax = sns.heatmap(
        corr,
        vmax=0.4,
        square=True,
        fmt=".1f",
        annot=True,
    )
```

# Are instrumental tracks more likely to be live recordings compared to vocal tracks?
No, opposite is True.

```python
dfcat = df.copy()
dfcat["instrumentalness"] = np.where(dfcat["instrumentalness"] > 0.70, 1, 0)
dfcat["speechiness"] = np.where(dfcat["speechiness"] > 0.70, 1, 0)
dfcat["liveness"] = np.where(dfcat["speechiness"] > 0.70, 1, 0)
```

```python
dfcat
```

```python
voclive = dfcat[(dfcat["speechiness"] == 1) & (dfcat["liveness"] == 1)].shape[0]
instlive = dfcat[(dfcat["instrumentalness"] == 1) & (dfcat["liveness"] == 1)].shape[0]
complist = [voclive, instlive]
namlist = ["Vocal+Live", "Instrumental+Live"]
```

```python
plt.bar(namlist, complist)
plt.ylabel('Instances')
plt.title('Vocal Instrumental comparision')
```
